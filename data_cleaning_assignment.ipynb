```python
import pandas as pd
import numpy as np

data = {
    "From_To": ["LoNDon_paris", "MAdrid_miLAN", "londON_StockhOlm", "Budapest_PaRis", "Brussels_londOn"],
    "FlightNumber": ["10045", "np.nan", "10065", "np.nan", "10085"],
    "RecentDelays": [[23, 47],[],[24, 43, 87],[13],[67, 32]],
    "Airline": ["KLM(!)", "<Air France> (12)", "(British Airways. )", "12. Air France", "Swiss Air"]
}
df = pd.DataFrame(data)
df
```

# Case:1

```python
df["FlightNumber"] = pd.to_numeric(df["FlightNumber"], errors='coerce')
df = df.replace(np.nan, 0, regex=True)
df["FlightNumber"] = df["FlightNumber"].astype(int)
for count, flight in enumerate(df["FlightNumber"]):
    if flight == 0:
        df["FlightNumber"][count] = df["FlightNumber"][count-1]+10
        
df
```

# Case:2

```python
from copy import deepcopy
city1 = []
city2 = []
for i in df['From_To']:
    cities = i.split("_")
    city1.append(cities[0])
    city2.append(cities[1])
    
df1 = deepcopy(df)
df1.insert(loc=0, column='FromCity', value=city1)
df1.insert(loc=1, column='ToCity', value=city2)
df1.drop(["From_To"], axis=1, inplace=True)
```

```python
df1
```

# Case:3

```python
df1['FromCity'] = df1['FromCity'].str.capitalize()
df1['ToCity'] = df1['ToCity'].str.capitalize()
df1
```

# Case:4

```python
df = df1
df
```

# Case:5

```python
delays_dict = {}
final_list = []
for i in df["RecentDelays"]:
    if len(i) == 3:
        final_list.append(i)
    elif len(i) == 2:
        i.insert(2, "NaN")
        final_list.append(i)
    elif len(i) == 1:
        i.insert(1, "NaN")
        i.insert(2, "NaN")
        final_list.append(i)
    else:
        final_list.append(["NaN","NaN","NaN"])

        
for j in range(3):
    l = []
    for k in range(5):
        l.append(final_list[k][j])
    delays_dict["Delay_"+str(j+1)] = l
          
df2 = pd.DataFrame(delays_dict)
```

```python
df3 = pd.concat([df1, df2], axis=1)
df3.drop(["RecentDelays"], axis=1, inplace=True)
df3
```
